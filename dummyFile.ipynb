```python
import os
import random

# 더미 파일을 생성할 기본 폴더 (BASE_DIR)
BASE_DIR = r"C:/Users/deank/Documents/test"  # 실제 사용 환경에 맞게 경로 수정

# 사용할 연도, 카테고리, 확장자 목록
years = ['2021', '2022', '2023', '2024', '2025']
categories = ['donation_receipt', '수입', '지출', 'Sermon', 'certificate', '마태복음', 'donateReceipt', '선교비-수입', '헌금  지출', '헌금, 수입', '12월 지출', 'camper']
extensions = ['pdf', 'docx', 'txt', 'doc', 'xlsx', 'csv']

# BASE_DIR 폴더가 없으면 생성
os.makedirs(BASE_DIR, exist_ok=True)

# 50개의 더미 파일 생성
num_files = 100
for i in range(1, num_files + 1):
    # 무작위로 연도, 카테고리, 확장자 선택
    year = random.choice(years)
    category = random.choice(categories)
    ext = random.choice(extensions)
    
    # 파일명 구성: 예) 2023_기부영수증_테스트_1.pdf
    filename = f"{category}_{year}-테스트_{i}.{ext}"
    file_path = os.path.join(BASE_DIR, filename)
    
    # 더미 파일 내용 작성
    with open(file_path, "w", encoding="utf-8") as f:
        f.write("테스트를 위한 더미 파일입니다.\n")
    
    print(f"더미 파일 생성 완료: {file_path}")
```
